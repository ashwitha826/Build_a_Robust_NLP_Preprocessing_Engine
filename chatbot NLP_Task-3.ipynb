# Load Model and Tokenizer :

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_name="Qwen/Qwen2-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model=AutoModelForCausalLM.from_pretrained(model_name,torch_dtype=torch.float32,device_map="auto")
```

```python
print("Chatbot : Hello ! I am your AI Assistant. Type 'exit', 'quit', or 'bye' to stop.")

while True:

  user_input=input("User : ")

  if user_input.lower() in ['exit','quit','bye']:
    print("Chatbot : Bye, Have a pleasurable day ahead!!! ")
    break

  prompt=f"""
  You are a helpful and knowledge base AI Assistant.
  Answer clearly and correctly in 2-3 sentences.

  Question : {user_input}
  Answer :
  """

  inputs=tokenizer(prompt,return_tensors="pt").to(model.device)

  outputs=model.generate(**inputs,max_new_tokens=100,do_sample=False,num_beams=3)

  response=tokenizer.decode(outputs[0],skip_special_tokens=True)

  print("Chatbot : ",response)
```
